```python
import mwclient

# On local, a descriptive User-Agent is usually enough
ua = 'MyTolkienBot/1.0 (contact: walidabdela98@gmail.com)'

site = mwclient.Site(
    'tolkiengateway.net',
    path='/w/',
    clients_useragent=ua
)

# This will likely work immediately on your local machine
page = site.pages['Frodo Baggins']
print(page.text())
```

```python
from rdflib import Namespace, URIRef, Literal, Graph
from rdflib.namespace import RDF, RDFS

# Define your custom namespaces
EX = Namespace("http://example.org/tolkien/")       # For the real entities [cite: 86]
PAGE = Namespace("http://example.org/tolkien/page/") # For the wiki pages [cite: 86]
SCHEMA = Namespace("http://schema.org/") # Define SCHEMA explicitly

g = Graph()
g.bind("ex", EX)
g.bind("schema", SCHEMA)
```

```python
import mwparserfromhell

def process_character(page_title):
    page = site.pages[page_title]
    text = page.text() # Get the source wikitext [cite: 91]

    # Use the 'resource' URI for the character entity [cite: 86]
    char_uri = EX[page_title.replace(" ", "_")]
    # Use the 'page' URI for the wiki article [cite: 86]
    page_uri = PAGE[page_title.replace(" ", "_")]

    # Create the link between page and entity [cite: 85]
    g.add((page_uri, SCHEMA.about, char_uri))
    g.add((char_uri, RDF.type, SCHEMA.Person)) # Example class [cite: 95]
    g.add((char_uri, RDFS.label, Literal(page_title, lang="en"))) [cite: 13, 89]

    # Parse the wikitext to find the infobox
    wikicode = mwparserfromhell.parse(text)
    templates = wikicode.filter_templates()

    for template in templates:
        if "Infobox character" in template.name.strip(): [cite: 76, 78]
            # Convert template parameters to triples [cite: 11, 79]
            if template.has("race"):
                race_val = template.get("race").value.strip()
                g.add((char_uri, EX.hasRace, Literal(race_val))) [cite: 11]
            if template.has("culture"):
                culture_val = template.get("culture").value.strip()
                g.add((char_uri, EX.hasCulture, Literal(culture_val))) [cite: 11]

# Test with Elrond [cite: 74, 77]
process_character("Elrond")
```
